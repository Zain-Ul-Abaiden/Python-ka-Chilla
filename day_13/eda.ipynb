```python
sns.countplot(data=df, x="pclass")
```

```python
# data distribution
sns.histplot(data=df, x='age', bins=15)
```

```python
df['age'].hist(bins=30, color='green', alpha=0.7)
```

```python
sns.boxplot(data=df, y='age', x='survived', hue='who')
# 1 side mai categorical or dosri side pai numeric values dene hai
```

```python
# scatterplot
sns.scatterplot(x='age', y='fare', data=df)
# dono side pai numeric data aye ga
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="survived")
```

```python
# stacked barchart / plott
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y='age', hue='who')
```

```python
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
# Relation of two numeric variable
sns.lmplot(data=df, x='age', y='fare')
```

```python
sns.lmplot( x='age', y='fare', data=df[df['fare'] < 300])
```

```python
sns.lmplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim(0,300)
```

```python
# how to add a column and make a age group
sns.histplot(data=df, x='age')
```

```python
# grouping
df['age_groups'] = pd.cut(df['age'], bins=[0, 10, 18, 30, 50, 80], labels=['kids', 'child', 'youth', 'middle ages', 'old'])
```

```python
df.head()
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```

```python
df['family_size']=pd.cut(df['parch']+ df['sibsp'], [0,1,2,6], labels=['single','couple','parents'])
```

```python
df.head()
```

```python
df.groupby('family_size')['survived'].mean().plot(kind='bar')
```
