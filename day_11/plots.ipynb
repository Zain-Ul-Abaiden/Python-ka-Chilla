# Plotting in python

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
# to check non values
kashti.isnull().sum()
```

# Barplot

```python
kashti.survived.value_counts().plot(kind="bar")
plt.title("Barplot of survived")
```

# countplot

```python
sns.countplot(x="survived", data=kashti)
plt.title("countplot of survived")
```

```python
sns.countplot(x="survived", hue="sex", data=kashti)
plt.title("countplot of survived with hue")
```

# Scatter plot

```python
plt.scatter(data=kashti, x="age", y="fare", c="pclass")
plt.title("scatterplot of age and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

```python
sns.scatterplot(data=kashti, x="age", y="fare", hue="pclass")
plt.title("scatterplot of age and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

```python
sns.barplot(data=kashti, x="sex", y="fare", hue="pclass")
plt.title("barplot of sex and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

# Pairplot

```python
sns.pairplot(kashti)
plt.title("Pairplot")
```

# histogram

```python
kashti.hist(bins=10, figsize=(10,10), grid=False)
plt.title("Histogram")
```

# Boxplot

```python
sns.boxplot(data=kashti, x="sex", y="fare", hue="parch")
plt.title("boxplot of sex and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

```python
sns.boxenplot(data=kashti, x="sex", y="fare", hue="class")
plt.title("boxplot of sex and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

# grid plot

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True, palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
plt.title("grid plot")
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="sex", margin_titles=True, palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True, palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

# stripplot

```python
sns.stripplot(data=kashti, x="sex", y="fare", hue="class")
plt.title("stripplot of sex and fare")
plt.xlabel("age")
plt.ylabel("fare")
```

# heatmap

```python
numeric_columns = kashti.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = kashti[numeric_columns].corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

# violinplot

```python
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")
plt.title("violinplot of sex and fare")
plt.xlabel("age")
plt.ylabel("fare")
```
