#16- Splitting a string into multiple columns

```python
import pandas as pd
df = pd.DataFrame({'name':['Zain Ul Abaiden', 'Muhammad Waseem Akram', 'Junaid Akram', 'Jamshed Akram'],
                   'location':['Lahore, Pakistan', 'Kasur, Pakistan', 'Islamabad, Pakistan', 'Karachi, Pakistan']})
df
```

```python
df[['first_name','middle_name', 'last_name']] = df.name.str.split(' ', expand=True)
```

```python
df
```

```python
df[['city_name','country_name']] = df.location.str.split(', ', expand=True)
df
```

```python
df = df[['first_name','middle_name', 'last_name', 'city_name', 'country_name']]
df
```

# 23- Reshaping a dataframe

```python
fasla = pd.DataFrame([['1234', 100, 200, 300], ['34567', 400, 500, 600], ['67890', 700, 800 ,900]], columns = ['zip', 'factory', 'warehouse', 'retail'])
fasla.head()
```

```python
fasla.head().T
```

```python
fasla2 = pd.DataFrame([[1 , '1234', 'factory'], [2, '34567', 'warehouse']], columns = ['user_id', 'zip', 'location_type'])
fasla2.head()
```

```python
fasla_long = fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
fasla_long.head()
```

```python
import seaborn as sns
sns.barplot(x='zip', y='distance', hue='location_type', data=fasla_long)
```
